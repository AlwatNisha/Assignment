```python
#healthcare industry:
class ClinicAppointment:
    def __init__(self):
        # Predefined doctors and time slots
        self.time_slots = ["10am", "11am", "12pm", "2pm", "3pm"]
        self.doctors = ["Dr. Sharma", "Dr. Mehta"]

        # Appointments data structure
        self.appointments = {
            doctor: {slot: [] for slot in self.time_slots} for doctor in self.doctors
        }

    def book_appointment(self):
        name = input("Enter patient name: ")
        age = input("Enter age: ")
        mobile = input("Enter mobile number: ")

        # Choose doctor
        print("\nAvailable Doctors:")
        for i, doc in enumerate(self.doctors, 1):
            print(f"{i}. {doc}")
        doc_choice = int(input("Choose doctor (number): ")) - 1
        doctor = self.doctors[doc_choice]

        # Choose time slot
        print("\nAvailable Time Slots:")
        for i, slot in enumerate(self.time_slots, 1):
            print(f"{i}. {slot} ({len(self.appointments[doctor][slot])}/3 booked)")
        slot_choice = int(input("Choose time slot (number): ")) - 1
        slot = self.time_slots[slot_choice]

        # Check availability
        if len(self.appointments[doctor][slot]) < 3:
            self.appointments[doctor][slot].append({
                "name": name,
                "age": age,
                "mobile": mobile
            })
            print(f"\n Appointment booked with {doctor} at {slot}.")
        else:
            print(f"\n Sorry, {slot} with {doctor} is fully booked.")

    def view_appointment(self):
        mobile = input("Enter your mobile number to view appointment: ")
        found = False
        for doctor in self.appointments:
            for slot in self.appointments[doctor]:
                for patient in self.appointments[doctor][slot]:
                    if patient["mobile"] == mobile:
                        print(f"\n Appointment Details:")
                        print(f"Name: {patient['name']}")
                        print(f"Age: {patient['age']}")
                        print(f"Doctor: {doctor}")
                        print(f"Time Slot: {slot}")
                        found = True
        if not found:
            print("\n No appointment found for this mobile number.")

    def cancel_appointment(self):
        mobile = input("Enter your mobile number to cancel appointment: ")
        found = False
        for doctor in self.appointments:
            for slot in self.appointments[doctor]:
                for patient in self.appointments[doctor][slot]:
                    if patient["mobile"] == mobile:
                        self.appointments[doctor][slot].remove(patient)
                        print(f"\n Appointment with {doctor} at {slot} cancelled.")
                        found = True
                        break
        if not found:
            print("\n No appointment found for this mobile number.")

    def run(self):
        while True:
            print("\n--- Clinic Appointment System ---")
            print("1. Book Appointment")
            print("2. View Appointment")
            print("3. Cancel Appointment")
            print("4. Exit")
            choice = input("Choose an option: ")

            if choice == "1":
                self.book_appointment()
            elif choice == "2":
                self.view_appointment()
            elif choice == "3":
                self.cancel_appointment()
            elif choice == "4":
                print("Goodbye!")
                break
            else:
                print(" Invalid choice. Please try again.")


# Run the system
clinic = ClinicAppointment()
clinic.run()
```

```python
#SchoolManagement 
class SchoolManagement:
    def __init__(self):
        self.students = {}  
        self.next_id = 1    

    def new_admission(self):
        name = input("Enter student name: ")
        age = int(input("Enter age: "))
        if age < 5 or age > 18:
            print("Age must be between 5 and 18.")
            return

        student_class = int(input("Enter class (1-12): "))
        if student_class < 1 or student_class > 12:
            print("Class must be between 1 and 12.")
            return

        mobile = input("Enter guardian's mobile number: ")
        if not (mobile.isdigit() and len(mobile) == 10):
            print("Mobile number must be 10 digits.")
            return

        student_id = self.next_id
        self.students[student_id] = {
            "name": name,
            "age": age,
            "class": student_class,
            "mobile": mobile
        }
        self.next_id += 1
        print(f"Student admitted! Assigned ID: {student_id}")

    def view_student(self):
        sid = int(input("Enter student ID: "))
        if sid in self.students:
            print("Student Details:")
            for key, value in self.students[sid].items():
                print(f"{key.capitalize()}: {value}")
        else:
            print("Student ID not found.")

    def update_student(self):
        sid = int(input("Enter student ID to update: "))
        if sid not in self.students:
            print("Student ID not found.")
            return

        print("1. Update Mobile Number")
        print("2. Update Class")
        choice = input("Choose option: ")

        if choice == "1":
            new_mobile = input("Enter new mobile number: ")
            if new_mobile.isdigit() and len(new_mobile) == 10:
                self.students[sid]["mobile"] = new_mobile
                print("Mobile number updated.")
            else:
                print("Invalid mobile number.")
        elif choice == "2":
            new_class = int(input("Enter new class (1-12): "))
            if 1 <= new_class <= 12:
                self.students[sid]["class"] = new_class
                print("Class updated.")
            else:
                print("Invalid class.")
        else:
            print("Invalid choice.")

    def remove_student(self):
        sid = int(input("Enter student ID to remove: "))
        if sid in self.students:
            del self.students[sid]
            print("Student record removed.")
        else:
            print("Student ID not found.")

    def run(self):
        while True:
            print("\n--- School Management System ---")
            print("1. New Admission")
            print("2. View Student Details")
            print("3. Update Student Info")
            print("4. Remove Student Record")
            print("5. Exit")
            choice = input("Choose option: ")

            if choice == "1":
                self.new_admission()
            elif choice == "2":
                self.view_student()
            elif choice == "3":
                self.update_student()
            elif choice == "4":
                self.remove_student()
            elif choice == "5":
                print("Goodbye!")
                break
            else:
                print("Invalid choice.")


# Run the program
school = SchoolManagement()
school.run()
```

```python
# transport reservation system(bus ticketing)
class BusReservation:
    def __init__(self):
        self.routes={ 
            "mumbai to pune":500,
            "delhi to jaipur":600,
            "ahemedabad to mumbai":400,
            "pune to nasik":500}
        self.tickets={}
        self.next_ticket_id=1
        self.seats_booked={route: 0 for route in self.routes} 

    def show_routes(self):
        print("\n--- Available Routes ---")
        for route, price in self.routes.items():
            print(f"{route} - â‚¹{price}")

    def book_ticket(self):
        self.show_routes()
        route = input("Enter route exactly as shown: ")
        if route not in self.routes:
            print("Route not found.")
            return

        if self.seats_booked[route] >= 40:
            print("No seats available for this route.")
            return

        name = input("Enter passenger name: ")
        age = int(input("Enter age: "))
        mobile = input("Enter mobile number: ")
        if not (mobile.isdigit() and len(mobile) == 10):
            print("Invalid mobile number.")
            return

        seat_no = self.seats_booked[route] + 1
        ticket_id = self.next_ticket_id
        self.next_ticket_id += 1

        self.tickets[ticket_id] = {
            "name": name,
            "age": age,
            "mobile": mobile,
            "route": route,
            "price": self.routes[route],
            "seat_no": seat_no
        }
        self.seats_booked[route] += 1

        print(f"Ticket booked! Ticket ID: {ticket_id}, Seat No: {seat_no}")

    def view_ticket(self):
        ticket_id = int(input("Enter ticket ID: "))
        if ticket_id in self.tickets:
            print("\n Ticket Details:")
            for key, value in self.tickets[ticket_id].items():
                print(f"{key.capitalize()}: {value}")
        else:
            print("Ticket not found.")

    def cancel_ticket(self):
        ticket_id = int(input("Enter ticket ID to cancel: "))
        if ticket_id in self.tickets:
            route = self.tickets[ticket_id]["route"]
            self.seats_booked[route] -= 1
            del self.tickets[ticket_id]
            print("Ticket cancelled successfully.")
        else:
            print("Ticket not found.")

    def run(self):
        while True:
            print("\n--- Bus Reservation System ---")
            print("1. Show Available Routes")
            print("2. Book Ticket")
            print("3. View Ticket")
            print("4. Cancel Ticket")
            print("5. Exit")

            choice = input("Choose option: ")

            if choice == "1":
                self.show_routes()
            elif choice == "2":
                self.book_ticket()
            elif choice == "3":
                self.view_ticket()
            elif choice == "4":
                self.cancel_ticket()
            elif choice == "5":
                print("Goodbye!")
                break
            else:
                print("Invalid choice.")


# Run the system
bus_system = BusReservation()
bus_system.run()                 
```
